```python
import pandas as pd
import matplotlib.pyplot as plt
from plotnine import *
import seaborn as sns
```

```python
databp = pd.read_csv("bp-stats-review-2019-consolidated-dataset-panel-format.csv")
```

```python
list_columns = ['Country', 'Year', 'pop', 'Region', 'co2_mtco2', 'coalcons_mtoe', 'coalprod_mtoe', 'gascons_mtoe', 'gasprod_mtoe', 'hydro_mtoe', 'hydro_twh', 'nuclear_mtoe',
    'nuclear_twh', 'oilcons_mtoe', 'oilprod_mt', 'primary_gj_pc', 'primary_mtoe', 'renewables_mtoe', 'renewables_twh', 'solar_mtoe', 'solar_twh', 'wind_mtoe', 'wind_twh']
```

```python
databp = databp[list_columns]
databp.Region.unique()
```

```python
databp = databp[databp['Country'].str.startswith('Total')]
databp = databp[databp['Country'].isin(['Total Africa', 'Total Europe', 'Total North America', 'Total S. & Cent. America', 'Total Asia Pacific', 'Total CIS', 'Total Middle East', 'Total World'])]
databp.Country.unique()
```

```python
table = databp[databp['Country'] != 'Total World']
```

```python
databp.tail()
```

```python
table.groupby('Year').sum().tail()
```

```python
databp
```

```python
databp['Year'] = databp['Year'].apply(str)
```

```python
databp['Year'] = pd.to_datetime(databp['Year'], format = '%Y')
```

```python
fig = plt.figure(figsize=(10,6))
sns.set(style='whitegrid',)
sns.lineplot(x="Year", y="pop", hue="Country", data=databp)
plt.title("Evolution of the population since 1965", fontsize = 16, fontweight='bold')
plt.xlabel("Year", fontsize = 15) 
plt.ylabel("Population (millions)", fontsize = 15) 
plt.legend(loc='upper right', prop={'size': 8})
plt.show()
```
